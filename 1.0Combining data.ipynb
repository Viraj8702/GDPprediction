## Business Confidence and Consumer Sentiment

```python
business_invest=pd.read_excel('data/business investment.xlsx')
business_invest.head(12)

# dropping the first 3 rows from the dataframe
business_invest=business_invest.drop(business_invest.index[:3])
business_invest
business_invest.rename(columns={'Unnamed: 1':'Gross Fixed Capital Formation','Unnamed: 2':'Business Investment'},inplace=True)
business_invest
```

```python
# Extract year from the 'Quarter' column
business_invest['Figure 1: Business investment has increased at a greater rate than gross fixed capital formation (GFCF) since 2022'] = business_invest['Figure 1: Business investment has increased at a greater rate than gross fixed capital formation (GFCF) since 2022'].str.extract(r'(\d{4})')[0]

# calculating the annual average
business_invest = business_invest.groupby('Figure 1: Business investment has increased at a greater rate than gross fixed capital formation (GFCF) since 2022').agg({
    'Gross Fixed Capital Formation': 'mean',
    'Business Investment': 'mean'
}).reset_index()

# Rename the columns for clarity
business_invest.columns = ['Year', 'Average GFCF', 'Average Business Investment']

# Print the results
print(business_invest)
```

```python
# dropping 2024 data
business_invest = business_invest.drop(index=24)
business_invest.tail()
```

## Combining all the datasetes into single dataframe

```python
# df = inflation_df.merge(cpih, on='Year')
# df = df.merge(gdp, on='Year')
```

```python
# inflation['Year'] = inflation['Year'].astype(int)
cpih['Year'] = cpih['Year'].astype(int)
gdp['Year'] = gdp['Year'].astype(int)
govern_spend['Year'] = govern_spend['Year'].astype(int)
household_consump['Year'] = household_consump['Year'].astype(int)
bankrate['Year'] = bankrate['Year'].astype(int)
fuel['Year'] = fuel['Year'].astype(int)
gas['Year'] = gas['Year'].astype(int)
trade_data_export['Year'] = trade_data_export['Year'].astype(int)
trade_data_import['Year'] = trade_data_import['Year'].astype(int)
business_invest['Year'] = business_invest['Year'].astype(int)
unemployment_rate['Year']=unemployment_rate['Year'].astype(int)
population_growth['Year'] = population_growth['Year'].astype(int)
```

```python
# print(inflation['Year'].dtype)
print(cpih['Year'].dtype)
print(gdp['Year'].dtype)
print(govern_spend['Year'].dtype)
print(household_consump['Year'].dtype)
print(bankrate['Year'].dtype)
print(fuel['Year'].dtype)
print(gas['Year'].dtype)
print(trade_data_export['Year'].dtype)
print(trade_data_import['Year'].dtype)
print(business_invest['Year'].dtype)
print(unemployment_rate['Year'].dtype)
print(population_growth['Year'].dtype)
```

```python
# Start by merging a couple, then add others progressively
# all_df = inflation.merge(cpih, on='Year')
# all_df = all_df.merge(gdp, on='Year')
all_df = cpih.merge(population_growth,on='Year')
all_df = all_df.merge(govern_spend, on='Year')
all_df = all_df.merge(household_consump, on='Year')
all_df = all_df.merge(bankrate, on='Year')
all_df = all_df.merge(fuel, on='Year')
all_df = all_df.merge(gas, on='Year')
all_df = all_df.merge(trade_data_export, on='Year')
all_df = all_df.merge(trade_data_import,on='Year')
all_df = all_df.merge(business_invest, on='Year')
all_df = all_df.merge(unemployment_rate, on='Year')
all_df = all_df.merge(gdp, on='Year')



# Check the combined dataframe
print(all_df.head())

# Save the combined dataframe to CSV (optional)
all_df.to_csv('data/allCombined_data.csv', index=False)
```
